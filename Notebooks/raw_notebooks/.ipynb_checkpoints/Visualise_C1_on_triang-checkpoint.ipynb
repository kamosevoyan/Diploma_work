# Interploation example

```python
t = np.linspace(np.pi/4, 2*np.pi+np.pi/4,5, endpoint=False)
points = np.stack((np.cos(t), np.sin(t))).T
seg = np.array([[j, j+1] for j in range(points.shape[0]-1)] + [[points.shape[0] - 1, 0]])
```

```python
cndt = triangulate({"vertices":points, "segments":seg}, opts="Rpq10a0.8ne")
points = cndt["vertices"]
triangles = cndt["triangles"]
neighbours = cndt["neighbors"]
triangles_s = points[triangles]
```

```python
is_boundary_point = np.zeros(points.shape[0], dtype=bool)

is_set = np.zeros(points.shape[0], dtype=np.bool_)

for index, element in enumerate(triangles):
    
    if (neighbours[index] == -1).sum() == 2:
        
        is_boundary_point[element] = True
        
    elif (neighbours[index] == -1).sum() == 1:
        
        indices_ = np.where(neighbours[index] != -1)[0]
        is_boundary_point[element[indices_]] = True
```

```python
%matplotlib inline
fig, ax = plt.subplots(figsize=(10, 10))
ax = plt.subplot(111, aspect='equal')
tplot(ax, **cndt)
plt.show()
```

```python
tx = np.linspace(-1, 1, 110)
ty = np.linspace(-1, 1, 110)
```

```python
mx, my = np.meshgrid(tx, ty, indexing="ij")
mz = np.zeros_like(mx)
```

```python
test_fn = sympy.sin(x)*sympy.cos(y)

test_fn_x = sympy.diff(test_fn, x)
test_fn_y = sympy.diff(test_fn, y)
test_fn_xx = sympy.diff(test_fn, x, x)
test_fn_xy = sympy.diff(test_fn, x, y)
test_fn_yy = sympy.diff(test_fn, y, y)
```

```python
test_fn_vals = np.zeros((points.shape[0], 6))

for idx, point in enumerate(points):
    
    test_fn_vals[idx][0] = sympy.lambdify((x, y), test_fn)(*point)
    test_fn_vals[idx][1] = sympy.lambdify((x, y), test_fn_x)(*point)
    test_fn_vals[idx][2] = sympy.lambdify((x, y), test_fn_y)(*point)
    test_fn_vals[idx][3] = sympy.lambdify((x, y), test_fn_xx)(*point)
    test_fn_vals[idx][4] = sympy.lambdify((x, y), test_fn_xy)(*point)
    test_fn_vals[idx][5] = sympy.lambdify((x, y), test_fn_yy)(*point)
```

```python
points_ = np.stack((mx, my)).transpose(1, 2, 0)

arg = (x_1, x_2, x_3, y_1, y_2, y_3)

for p_idx, t_points in zip(triangles, triangles_s):
    
    
    a = t_points[0] - t_points[1]
    b = t_points[0] - t_points[2]
    
    areas = np.abs(a[0]*b[1] - a[1]*b[0])
        
    a = (points_[...,] - t_points[0, None, None])
    b = (points_[...,] - t_points[1, None, None])
    c = (points_[...,] - t_points[2, None, None])
    
    area1 = np.abs(a[..., 0] * b[..., 1] - a[..., 1] * b[..., 0])
    area2 = np.abs(a[..., 0] * c[..., 1] - a[..., 1] * c[..., 0])
    area3 = np.abs(b[..., 0] * c[..., 1] - b[..., 1] * c[..., 0])
    indices = np.where(np.isclose((area1+area2+area3), areas))
    
    
    CURRENT_BASIS = BELL_BASIS.map_to_cell(t_points)
 
    I = 5
    
    on_1 =  CURRENT_BASIS[0+I].as_sympy() * test_fn_vals[p_idx[0]][I]
    on_2 =  CURRENT_BASIS[6+I].as_sympy() * test_fn_vals[p_idx[1]][I]
    on_3 = CURRENT_BASIS[12+I].as_sympy() * test_fn_vals[p_idx[2]][I]
    
    
    F = sympy.lambdify((x, y), on_1 + on_2 + on_3)
    
    
    MZ = F(mx, my)
    
    if isinstance(MZ, (float, int)):
        print(on_2)
        
        MZ = np.ones_like(mx) * MZ
    else:
        print("B")
    
    mz[indices] = MZ[indices]
```

```python
%matplotlib widget
fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection='3d')
ax.plot_surface(mx, my, mz)
# ax.plot_surface(mx, my, sympy.lambdify((x, y), test_fn)(mx, my))
plt.show()
```

```python
points_ = np.stack((mx, my)).transpose(1, 2, 0)

on_1 = bell_basis[0:6]
on_2 = bell_basis[6:12]
on_3 = bell_basis[12:18]


arg = (x_1, x_2, x_3, y_1, y_2, y_3)

for p_idx, t_points in zip(triangles, triangles_s):
    
    
    a = t_points[0] - t_points[1]
    b = t_points[0] - t_points[2]
    
    areas = np.abs(a[0]*b[1] - a[1]*b[0])
    
    del a
    del b
    
    
    a = (points_[...,] - t_points[0, None, None])
    b = (points_[...,] - t_points[1, None, None])
    c = (points_[...,] - t_points[2, None, None])
    
    area1 = np.abs(a[..., 0] * b[..., 1] - a[..., 1] * b[..., 0])
    area2 = np.abs(a[..., 0] * c[..., 1] - a[..., 1] * c[..., 0])
    area3 = np.abs(b[..., 0] * c[..., 1] - b[..., 1] * c[..., 0])
    indices = np.where(np.isclose((area1+area2+area3), areas))
    
    del area1, area2, area3
    
    
    
    P_1_x = sympy.lambdify((x_1, x_2, x_3, y_1, y_2, y_3,), sympy.diff(P_1, x))(*t_points[:, 0], *t_points[:, 1])
    P_2_x = sympy.lambdify((x_1, x_2, x_3, y_1, y_2, y_3,), sympy.diff(P_2, x))(*t_points[:, 0], *t_points[:, 1])
        
    P_1_y = sympy.lambdify((x_1, x_2, x_3, y_1, y_2, y_3,), sympy.diff(P_1, y))(*t_points[:, 0], *t_points[:, 1])
    P_2_y = sympy.lambdify((x_1, x_2, x_3, y_1, y_2, y_3,), sympy.diff(P_2, y))(*t_points[:, 0], *t_points[:, 1])
    
    
#     W1 = test_fn_vals[p_idx[0]][0]
#     W2 = test_fn_vals[p_idx[1]][0]
#     W3 = test_fn_vals[p_idx[2]][0]
    
#     xy1 = on_1[0] * W1
#     xy2 = on_2[0] * W2
#     xy3 = on_3[0] * W3
    

    W1 = (test_fn_vals[p_idx[0]][1]/P_1_x-test_fn_vals[p_idx[0]][2]/P_1_y)/(P_2_x/P_1_x-P_2_y/P_1_y)
    W2 = (test_fn_vals[p_idx[1]][1]/P_1_x-test_fn_vals[p_idx[1]][2]/P_1_y)/(P_2_x/P_1_x-P_2_y/P_1_y)
    W3 = (test_fn_vals[p_idx[2]][1]/P_1_x-test_fn_vals[p_idx[2]][2]/P_1_y)/(P_2_x/P_1_x-P_2_y/P_1_y)
    
    xy1 = on_1[2] * W1
    xy2 = on_2[2] * W2
    xy3 = on_3[2] * W3
        
    
    F = sympy.lambdify((p_1, p_2), xy1+xy2+xy3)
    
    P_1_m = sympy.lambdify((x_1, x_2, x_3, y_1, y_2, y_3, x, y), P_1)(*t_points[:, 0], *t_points[:, 1], mx[indices], my[indices])
    P_2_m = sympy.lambdify((x_1, x_2, x_3, y_1, y_2, y_3, x, y), P_2)(*t_points[:, 0], *t_points[:, 1], mx[indices], my[indices])

    
    MZ = F(P_1_m, P_2_m)
    
    mz[indices] = MZ
```

```python
%matplotlib widget
fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection='3d')
ax.plot_surface(mx, my, mz)
# ax.plot_surface(mx, my, sympy.lambdify((x, y), test_fn)(mx, my))
plt.show()
```
