```python
import torch

class MyModule(torch.nn.Module):

    def __init__(self):
        super(MyModule, self).__init__()
        self.l1 = torch.nn.Linear(10, 5)
        self.l2 = torch.nn.Linear(5, 2)
        self.sf = torch.nn.Softmax(-1)
        

    def forward(self, x):
        return self.sf(self.l2(self.l1(x)))

# Compile the model code to a static representation
my_script_module = torch.jit.script(MyModule())

# Save the compiled code and model data so it can be loaded elsewhere
my_script_module.save("my_script_module.pt")
```

```python
model = torch.jit.load("my_script_module.pt")
```

```python
model(torch.rand(100, 10))
```
